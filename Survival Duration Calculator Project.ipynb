```python
#Survival Duration Calculator Project

# Objective- The objective of the project is to calculate the duration of how
# long a person lived based on his age. The duration should be calculated in
# different time units like Months, Weeks, Days, Hours, Minutes, Seconds.
# The input of the application is the age of the person and the output is expected to calculate the duration of time he lived
# in different time units as selected by the user.



def age_to_duration(age, time_unit):
    # Conversion factors for different time units
    conversion_factors = {
        'months': 12,
        'weeks': 52.1775,
        'days': 365.25,
        'hours': 365.25 * 24,
        'minutes': 365.25 * 24 * 60,
        'seconds': 365.25 * 24 * 60 * 60,
    }

    if time_unit in conversion_factors:
        duration = age * conversion_factors[time_unit]
        return duration
    else:
        return "Invalid time unit"

# Example usage:
age = float(input("Enter the age of the person: "))
time_unit = input("Enter the desired time unit (e.g., months, weeks, days, hours, minutes, seconds): ")

result = age_to_duration(age, time_unit)
if result != "Invalid time unit":
    print(f"The person has lived for {result:.2f} {time_unit}.")
else:
    print("Invalid time unit entered.")
```
